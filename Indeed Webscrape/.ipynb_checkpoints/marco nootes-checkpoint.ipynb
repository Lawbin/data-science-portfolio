```python
import numpy as np
URL ='https://hk.indeed.com'
page='/jobs?q=data+analyst&sort=date&start='
titles =[]
locations=[]
companies=[]
dess=[]
for i in range(85):
    morelink = URL + page
    y = str(i*10)
    morelink1 = morelink+ y
    subhtml = requests.get(morelink1)
    soup = BeautifulSoup(subhtml.text,'html.parser')
   
    for i in range(len(soup.find_all(class_ = 'title'))):
        try:
            title = soup.find_all(class_ = 'title')[i].text
            titles.append(title)
        except:
            title = np.nan
            titles.append(title)
        try:
            location = soup.find_all(class_='location accessible-contrast-color-location')[i].text
            locations.append(location)
        except:
            location=np.nan
            locations.append(location)
        try:
            company=soup.find_all(class_='company')[i].text
            companies.append(company)
        except:
            company=np.nan
            companies.append(company)
            
        jobresp = soup.find_all(class_="jobtitle turnstileLink")[i]['href']
        jobresp=URL+jobresp
        if jobresp:
            subhtml=requests.get(jobresp)
            sub=BeautifulSoup(subhtml.text,'html.parser')
            jobdes=sub.find_all(class_="jobsearch-jobDescriptionText")[0].text
            dess.append(jobdes)
    #morelink = URL + page
    #y = str(i*10)
    #morelink1 = morelink+ y
    #subhtml = requests.get(morelink1)
    #print(morelink1)
    #soup = BeautifulSoup(subhtml.text,'html.parser')
```

```python
import pandas as pd
#l = [titles,locations,companies,dess]
d = {'job title':titles,'location':locations,'company':companies,'job description':dess}
#columns = ['job title','location','company','job description']
indeed = pd.DataFrame(d)
indeed
```
